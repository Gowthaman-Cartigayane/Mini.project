```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('games')
cursor = conn.cursor()
```

```python
games_df= pd.read_csv("games.csv")
vgsales_df = pd.read_csv("vgsales.csv")
```

```python
vgsales_df
```

```python
def clean_title(title):
  if pd.isnull(title):
    return ''
  return str(title).strip().lower().replace(":", "").replace("-", "").replace("," ,"").replace("." , "")

games_df['clean_title']= games_df['Title'].apply(clean_title)
vgsales_df['clean_name']=vgsales_df['Name'].apply(clean_title)

merged_df= pd.merge(games_df, vgsales_df, how='inner',  left_on='clean_title', right_on='clean_name')

merged_df.drop(columns=['clean_title', 'clean_name'], inplace = True)

merged_df.to_csv("merged_gaames_data.csv", index = False)

print("Merging complete, Final shape:", merged_df.shape)
```

Question 28

```python
genre_engagement = merged_df.groupby('Genre').agg({
    'Wishlist': 'mean',
    'Global_Sales': 'mean',
    'Rating': 'mean'
}).reset_index()
genre_engagement_melted = genre_engagement.melt(
    id_vars='Genre',
    value_vars=['Wishlist', 'Global_Sales', 'Rating'],
    var_name='Metric',
    value_name='Value')
plt.figure(figsize=(14,6))
sns.barplot(data=genre_engagement_melted, x='Genre', y='Value', hue='Metric', palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title('User Engagement Metrics Across Genres')
plt.ylabel('Average Value')
plt.xlabel('Genre')
plt.legend(title='Metric')
plt.tight_layout()
plt.show()
```
